## Ejercicio 1. Definición de la clase Cancion
1.1. Define  la clase Cancion, la cual va a tener 4 atributos básicos. (título,  artista, album, genero y duración)

1.2. Crea un método para la clase Cancion que nos muestre la información de la canción.

1.3. Crea un método que nos permita comparar la duración de dos canciones y nos indique cuál de las dos es más larga.

1.4. Crea un método que nos permita calcular la popularidad de una canción en función de varios criterios.

- La puntuación base debe ser de 50.
    
- Las canciones de entre 2 y 4 minutos de duración son más populares, por lo que suman 20 puntos. Las demás restan 10.  

- Las canciones de los géneros pop, rock, hip-hop y electrónicas son populares, por lo que suman 20 puntos. El resto de géneros restan 5 puntos.

- Si una canción pertenece a un álbum suma 10 puntos.

1.5. Crea dos instancias de la clase canción con la siguiente info: ("Stairway to Heaven", "Led Zeppelin", "Led Zeppelin IV", 8.02, "Rock") y ("What's Going On", "Marvin Gaye", "What's Going On", 3.53, "Soul")

1.6. Comprueba el funcionamiento de cada uno de los métodos de la clase



````python
class Cancion:
  #1.1
  def __init__(self, titulo, artista, album, genero, duracion):
    self.titulo = titulo
    self.artista = artista
    self.album = album
    self.genero = genero
    self.duracion = duracion

  #1.2
  def info(self):
      return (f"Título: {self.titulo}\n"
              f"Artista: {self.artista}\n"
              f"Álbum: {self.album}\n"
              f"Género: {self.genero}\n"
              f"Duración: {self.duracion} segundos\n")
  
  #1.3
  def comparador(self, nueva_cancion):
      if self.duracion > nueva_cancion.duracion:
        return f"La cancion '{self.titulo}' es mas larga que la cancion '{nueva_cancion.titulo}'"
      elif self.duracion > nueva_cancion.duracion:
         return f"La cancion '{self.titulo}' es mas larga que la cancion '{nueva_cancion.titulo}'"
      else:
         return f"Ambas canciones son iguales"
      
  
  #1.4
  def puntuacion(self, lista_popular):
      puntuacion_base = 50

      if self.album != None:
         puntuacion_base = puntuacion_base + 10
      if int(self.duracion) > 120 & int(self.duracion) < 240:
         puntuacion_base = puntuacion_base + 20
      if int(self.duracion) < 120 or int(self.duracion) > 240:
         puntuacion_base = puntuacion_base -10
      if self.genero in lista_popular:
         puntuacion_base = puntuacion_base + 20
      if self.genero not in lista_popular:
         puntuacion_base = puntuacion_base - 5
      return puntuacion_base
      

#1.5 He usado una instancia personalizada
cancion1 = Cancion("Llovera", "Ramon", "Al otro lado de la estacion", "ñoño", "150")
nueva_cancion = Cancion("Bohemian Rapsodi", "Queen", "El que sea", "rock", "480")
#1.6 He comprobado el funcionamiento de la clase y los metodos en estas instancias
lista_popular = ['pop','rock','hip-hop','electrónicas']
print(cancion1.info())
print(cancion1.comparador(nueva_cancion))
print(cancion1.puntuacion(lista_popular))
print(nueva_cancion.puntuacion(lista_popular))
````

## Ejercicio 2. Definición de la subclase Colaboracion

````python
# 2.1 https://www.geeksforgeeks.org/how-to-make-a-subclass-from-a-super-class-in-python/
class SubColaboracion(Cancion):
    def __init__(self, titulo, artista, album, genero, duracion, artistas_colaboradores):
        super().__init__(titulo, artista, album, genero, duracion)
        self.artistas_colaboradores = artistas_colaboradores
# 2.2
    def subinfo(self):
      return (f"Título: {self.titulo}\n"
              f"Artista: {self.artista}\n"
              f"Álbum: {self.album}\n"
              f"Género: {self.genero}\n"
              f"Duración: {self.duracion} segundos\n"
              f"Artistas Colaboradores: {self.artistas_colaboradores} como colaborador(es)\n")
 # 2.3, 2.4 y 2.6 aqui tambien compruebo si hay uno o mas colaboaradores ademas de su numero. Tambien saco el colaborador principal.
    def comprobar_colaboracion(self):
       if self.artistas_colaboradores != None:
          if isinstance(self.artistas_colaboradores, list):
            if len(self.artistas_colaboradores) > 1:
               return f"La cancion {self.titulo} es una colaboracion y hay {len(self.artistas_colaboradores)} colaboradores que son {', '.join(self.artistas_colaboradores)}.\n El colaborador principal es {self.artistas_colaboradores[0]}"  # haciendo el mensaje mas claro
            else:
               return f"La cancion {self.titulo} es una colaboracion y el colaborador unico y principal es {self.artistas_colaboradores[0]}" # haciendo el mensaje mas claro
       else:
            return f"La cancion {self.titulo} no es una colaboracion"
       
# 2.5
    def puntuacionColaboracion(self, lista_popular):
      puntuacion_base = 50
      if self.album != None:
         puntuacion_base = puntuacion_base + 10
      if int(self.duracion) > 120 and int(self.duracion) < 240:
         puntuacion_base = puntuacion_base + 20
      if int(self.duracion) < 120 or int(self.duracion) > 240:
         puntuacion_base = puntuacion_base -10
      if self.genero in lista_popular:
         puntuacion_base = puntuacion_base + 20
      if self.genero not in lista_popular:
         puntuacion_base = puntuacion_base - 5
      if len(self.artistas_colaboradores) > 0:
         puntuacion_base = puntuacion_base + len(self.artistas_colaboradores) * 5
      return puntuacion_base
          
# 2.7 he usado mis propias instancias
cancion2 = SubColaboracion("Llovera", "Ramon", "Al otro lado de la estacion", "ñoño", "150",["Presuntos implicados", "Mojinos Escozios"])
cancion3 = SubColaboracion("Llovera", "Ramon", "Al otro lado de la estacion", "ñoño", "150",["Fito"])
# 2.8 Comprobacion de los metodos usando dos instancias      
print(cancion2.subinfo())
print(cancion2.comprobar_colaboracion())
print(cancion3.comprobar_colaboracion())
print(cancion2.puntuacionColaboracion(lista_popular))
print(cancion3.puntuacionColaboracion(lista_popular))
````

````python
# 3.1
class CancionPremiada(Cancion):
    def __init__(self, titulo, artista, album, genero, duracion, premios):
        super().__init__(titulo, artista, album, genero, duracion)
        self.premios = premios
    
    # 3.2
    def subinfopremios(self):
      return (f"Título: {self.titulo}\n"
              f"Artista: {self.artista}\n"
              f"Álbum: {self.album}\n"
              f"Género: {self.genero}\n"
              f"Duración: {self.duracion} segundos\n"
              f"Premios: {self.premios} premios\n")
    
    # 3.3
    def numeroPremios(self):
        if len(self.premios) > 1:
            return f"Este temazo ha ganado los siguientes premios: {', '.join(self.premios)}"
        if len(self.premios) == 1:
            return f"Este temazo ha ganado un premio: {self.premios[0]}"
        if len(self.premios) == 0:
            return f"Esta cancion no se ha llevado ningun premio"
             
    # 3.4
    def numeroPremiosPrestige(self, lista):
        premios_prestigiosos = []
        if len(self.premios) > 1:
            for l in self.premios:
                if l in lista:
                    premios_prestigiosos.append(l)
                return f"Este temazo ha ganado los siguientes premios prestigosos: {', '.join(premios_prestigiosos)}"
        if len(self.premios) == 0:
            return f"Esta cancion no se ha llevado ningun premio"

# 3.5 Creo mi propia instancia
lista_de_premios_prestigiosos = ["Grammy", "Oscar", "Golder Globe", "Brit Award"]
cancion3 = CancionPremiada("Llovera", "Ramon", "Al otro lado de la estacion", "Pop", "150",["Grammy", "American Music Award", "Mejor Cancion Fiestas del Bando de la Huerta 1988"])
# 3.6 Compruebo que todo funciona
print(cancion3.subinfopremios())
print(cancion3.numeroPremios())
print(cancion3.numeroPremiosPrestige(lista_de_premios_prestigiosos))
````
